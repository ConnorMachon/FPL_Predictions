```python
import pandas as pd
import requests
import numpy as np
```

# Accessing the FPL API & Reading In the Data

```python
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
r = requests.get(url)
json = r.json()
```

# Building Our 3 Dataframes (Elements, Element_Types, & Teams)

```python
# uncommmenting the next line lets you see what dictionary keys are included in the json data
# json.keys()
elements_df = pd.DataFrame(json['elements'])
### This is the MAIN dataframe and includes all of the data points that we care about
elements_types_df = pd.DataFrame(json['element_types'])
### This is a SUPPORTING dataframe and will help map position and other elements to the main df
teams_df = pd.DataFrame(json['teams'])
### This is a SUPPORTING dataframe and will help map teams to the players in the main df
```

```python
# we want to change the settings so that we can show all of the dataframes' columns
### Step 1) see how many columns each of the dataframes have (rows, columns)
elements_df.shape
### 67 columns
elements_types_df.shape
### 11 columns
teams_df.shape
### 21 columns

### Step 2) change pandas settings to display all columns (and rows just in case) for the df with the most columns
pd.set_option('display.max_columns', 67)
pd.set_option('display.max_rows', 67)
```
